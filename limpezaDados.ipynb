```python
import pandas as pd
```

```python
df = pd.read_csv("vendas_modificado (3).csv")
df.head()
```

```python
print(df.dtypes)
```

```python
df["valor"] = (
    df["valor"]
    .astype(str)
    .str.replace("R\$", "", regex=True)
    .str.replace(",", ".", regex=False)
    .str.strip()
)

df["valor"] = pd.to_numeric(df["valor"], errors="coerce")
```

```python
df.head()
```

```python
df["valor"] = pd.to_numeric(df["valor"], errors="coerce")
df["quantidade"] = pd.to_numeric(df["quantidade"], errors="coerce")
df["total"] = pd.to_numeric(df["total"], errors="coerce")
df["frete"] = pd.to_numeric(df["frete"], errors="coerce")

print(df.dtypes)
```

```python
df.head()
```

```python
print(df.isnull().sum())
```

```python
df.head()
```

```python
df["valor"] = df["valor"].fillna(df["valor"].mean())


df["valor"] = df["valor"].fillna(df["valor"].median())
```

```python
df.head()
```

```python
df["frete"] = df["frete"].fillna(df["frete"].median())
df["vendedor"] = df["vendedor"].fillna("Desconhecido")
```

```python
print(df[df["valor"].isna()])
print(df[df["total"].isna()])
```

```python
df['data'] = pd.to_datetime(df['data'], errors='coerce').dt.strftime('%Y-%m-%d')


df['hora'] = pd.to_datetime(df['hora'], errors='coerce').dt.strftime('%H:%M:%S')
```

```python
df['quantidade'] = df['quantidade'].astype(int)
```

```python
df['total_calculado'] = df['valor'] * df['quantidade'] + df['frete']
```

```python
df['total_incorreto'] = df['total'].round(2) != df['total_calculado'].round(2)
```

```python
df.head()
```

```python
df['total'] = df['total_calculado'].round(2)
```

```python
df.head()
```

```python
df.drop(columns=['total_calculado', 'total_incorreto'], inplace=True)
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.head()
```

```python
cat_cols = ['produto', 'status', 'pagamento', 'vendedor', 'marca', 'cidade', 'estado']
for col in cat_cols:
    df[col] = df[col].str.strip().str.lower()
```

```python
df = df[df['cep'] != '00000-000']
```

```python
df['data'] = pd.to_datetime(df['data'], errors='coerce')
df = df.dropna(subset=['data'])
```

```python
df.head()
```

```python
df['cliente'] = df['cliente'].str.replace(r'\n', ' ', regex=True).str.strip()
```

```python
df.head()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['frete'] = df['frete'].apply(lambda x: max(x, 0))
```

```python
import re
df['produto'] = df['produto'].apply(lambda x: re.sub(r'[^A-Za-zÀ-ÿ\s]', '', str(x)))
```

```python
df['cep'] = df['cep'].astype(str)
cep_valido = df['cep'].str.match(r'^\d{5}-\d{3}$')

df = df[cep_valido].copy()
```

```python
import re
```

```python
def normalizar_produto(texto):
    if pd.isnull(texto):
        return texto
    texto = str(texto).lower()
    texto = unicodedata.normalize('NFKD', texto).encode('ASCII', 'ignore').decode('utf-8')
    texto = re.sub(r'[^a-z0-9\s]', '', texto)
    return texto.strip()
```

```python
df['produto'] = df['produto'].apply(normalizar_produto)
```

```python
correcoes = {
    'refrigkrante': 'refrigerante',
    'scl': 'sal',
    'agua mineras': 'agua mineral',
    'presuntd': 'presunto'
}

df['produto'] = df['produto'].replace(correcoes)
```

```python
def normalizar_marca(texto):
    if pd.isnull(texto):
        return texto
    texto = str(texto).lower()
    texto = unicodedata.normalize('NFKD', texto).encode('ASCII', 'ignore').decode('utf-8')
    texto = texto.replace('-', ' ')
    texto = re.sub(r'[^a-z0-9\s]', '', texto)
    return texto.strip()

df['marca'] = df['marca'].apply(normalizar_marca)
```

```python
correcoes_marca = {
    'guarana antartica': 'guarana antarctica',
    'ki brasa': 'kibrasa'
}

df['marca'] = df['marca'].replace(correcoes_marca)
```
