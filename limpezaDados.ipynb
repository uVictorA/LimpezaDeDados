```python
import difflib

# Lista de produtos válidos
produtos_validos = [
    'cafe', 'biscoito recheado', 'papel higienico', 'macarrao', 'shampoo',
    'queijo mussarela', 'leite integral', 'sabao em po', 'arroz', 'pasta de dente',
    'feijao', 'agua mineral', 'presunto', 'refrigerante', 'pao de forma',
    'amaciante', 'manteiga', 'acucar', 'desinfetante', 'papel toalha',
    'oleo de soja', 'suco de laranja', 'vinho', 'sabonete', 'cerveja',
    'detergente', 'farinha de trigo', 'condicionador', 'molho de tomate'
]

# Função que corrige o nome do produto com base nos válidos
def corrigir_produto(nome, whitelist, cutoff=0.7):
    nome = str(nome).lower().strip()
    matches = difflib.get_close_matches(nome, whitelist, n=1, cutoff=cutoff)
    return matches[0] if matches else nome  # se achar parecido, retorna a correção; se não, mantém o original
```

```python
# Corrige os nomes da coluna "produto"
df['produto_corrigido'] = df['produto'].apply(lambda x: corrigir_produto(x, produtos_validos))
```

```python
produtos_corrigidos = df[df['produto'] != df['produto_corrigido']][['produto', 'produto_corrigido']]
print(produtos_corrigidos.drop_duplicates().head(20))
```
